```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
x_train = pd.DataFrame(data)
string_labels = ['Bipolar Type-2','Depression','Bipolar Type-1','Bipolar Type-2','Normal']
repetitions = len(x_train) // len(string_labels)
remainder = len(x_train) % len(string_labels)
y_train = string_labels * repetitions + string_labels[:remainder]
numeric_columns = x_train.select_dtypes(include=['int64', 'float64']).columns
categorical_columns = x_train.select_dtypes(include=['object']).columns
label_encoders = {}
for column in categorical_columns:
    le = LabelEncoder()
    x_train[column] = le.fit_transform(x_train[column])
    label_encoders[column] = le
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
new_data = {
    'Sadness': ['Sometimes'],
    'Euphoric': ['Sometimes'],
    'Exhausted': ['Usually'],
    'Sleep dissorder': ['Seldom'],
    'Mood Swing': ['YES'],
    'Suicidal thoughts':['NO'],
    'Anorxia': ['NO'],
    'Authority Respect': ['YES'],
    'Try-Explanation': ['YES'],
    'Aggressive Response': ['YES'],
    'Ignore & Move-On': ['NO'],
    'Nervous Break-down': ['NO'],
    'Admit Mistakes': ['YES'],
    'Overthinking': ['YES'],
    'Concentration': [100],
    'Optimism': [8],
    'Expert Diagnose': ['Normal']
}
input_df = pd.DataFrame(new_data)
for column in categorical_columns:
    input_df[column] = label_encoders[column].transform(input_df[column])
prediction = model.predict(input_df)


```
