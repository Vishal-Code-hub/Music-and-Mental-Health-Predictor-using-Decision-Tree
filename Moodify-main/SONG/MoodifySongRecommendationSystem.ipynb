```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier


# Create the training DataFrame
x_train = pd.DataFrame(data)
string_labels = ['Latin', 'Rock', 'Video game music', 'Jazz', 'R&B']
repetitions = len(x_train) // len(string_labels)
remainder = len(x_train) % len(string_labels)
y_train = string_labels * repetitions + string_labels[:remainder]  # Adjust to match length
 # Replace with your actual target values

# Separate numeric and categorical columns
numeric_columns = x_train.select_dtypes(include=['int64', 'float64']).columns
categorical_columns = x_train.select_dtypes(include=['object']).columns

# Initialize LabelEncoder and encode categorical columns
label_encoders = {}
for column in categorical_columns:
    le = LabelEncoder()
    x_train[column] = le.fit_transform(x_train[column])
    label_encoders[column] = le  # Store the encoder for later use

# Fit the Decision Tree model
model = DecisionTreeClassifier()
model.fit(x_train, y_train)

# Prepare input for prediction with string values
new_data = {
    'Age': [18],  # Example numeric input
    'Hours per day': [3],  # Example numeric input
    'Fav genre': ['Latin'],
    'Exploratory': ['Yes'],  # String value to encode
    'Foreign languages': ['Yes'],  # String value to encode
    'BPM':[156],
    'Frequency [Classical]': ['Rarely'],
    'Frequency [Country]': ['Never'],
    'Frequency [EDM]': ['Rarely'],
    'Frequency [Folk]': ['Never'],
    'Frequency [Gospel]': ['Never'],
    'Frequency [Hip hop]': ['Sometimes'],
    'Frequency [Jazz]': ['Never'],
    'Frequency [K pop]': ['Very frequently'],
    'Frequency [Latin]': ['Very frequently'],
    'Frequency [Lofi]': ['Rarely'],
    'Frequency [Metal]': ['Never'],
    'Frequency [Pop]': ['Very frequently'],
    'Frequency [R&B]': ['Sometimes'],
    'Frequency [Rap]': ['Very frequently'],
    'Frequency [Rock]': ['Never'],
    'Frequency [Video game music]': ['Sometimes'],
    'Anxiety': [3],
    'Depression': [0],
    'Insomnia': [1],
    'OCD': [0],
}

# Create DataFrame for the new input
input_df = pd.DataFrame(new_data)

# Encode categorical columns in the input DataFrame
for column in categorical_columns:
    input_df[column] = label_encoders[column].transform(input_df[column])

# Make prediction
prediction = model.predict(input_df)


```

```python
# Output the prediction
print("Prediction:", prediction)
```
